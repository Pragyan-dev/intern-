\# Data Analysis of Titanic Dataset

This notebook reads the Titanic dataset, performs basic data cleaning, and visualizes the passenger class distribution using matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
print("\nNull values before cleaning:")
print(df.isnull().sum())
```

```python
df_clean = df.dropna(subset=['Pclass', 'Sex', 'Embarked'])
```

```python
counts = df_clean['Pclass'].value_counts().sort_index()
```

```python
class_labels = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}
counts.index = [class_labels[i] for i in counts.index]
```

```python
plt.subplot(1, 2, 1)
counts.plot(kind='bar', color='lightblue', edgecolor='navy')
plt.title('Titanic Passenger Class Distribution (Matplotlib)')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
plt.subplot(1, 2, 2)
sns.barplot(x=counts.index, y=counts.values, palette='Set2')
plt.title('Titanic Passenger Class Distribution (Seaborn)')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
